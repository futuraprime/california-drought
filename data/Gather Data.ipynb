```python
import json
import requests
import csvkit
import StringIO
```

```python
with open('stations.json') as f:
    stations = json.loads(f.read())
stations
```

## The api link

http://cdec.water.ca.gov/cgi-progs/queryCSV?station_id=ALM&dur_code=M&sensor_num=15&start_date=1901/01/01

```python
csv_url = "http://cdec.water.ca.gov/cgi-progs/queryCSV"
```

```python
r = requests.get(csv_url, params={
    'station_id' : "ALM",
    'dur_code' : 'M',
    'sensor_num' : 15,
    'start_date' : '1901/01/01',
})
r.url
```

```python
r.text
```

```python
out = StringIO.StringIO(r.text)
reader = csvkit.reader(out)
dat = {}

# first two rows are headers
reader.next()
reader.next()

for r in reader:
    dat[r[0]] = r[2]

dat
```
