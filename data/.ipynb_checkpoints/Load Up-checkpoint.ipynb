```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import requests
from bs4 import BeautifulSoup as Soup
import re
```

```python
listings_url = "http://cdec.water.ca.gov/cgi-progs/staSearch?staid=&sensor_chk=on&sensor=15&dur=D&active=&lon1=&lon2=&lat1=&lat2=&elev1=-5&elev2=99000&nearby=&display=staid"
```

```python
reservoir_listing = Soup(requests.get(listings_url).content)
```

```python
rows = reservoir_listing.findAll('tr')
rows
```

```python
headings = [[h.text.strip() for h in row.findAll('th')] for row in rows if row.find('th')][0]
headings
```

```python
values = [[d.text for d in row.findAll('td')] for row in rows if row.find('td')]
values
```

```python
from titlecase import titlecase
```

```python
def cleanReservoirDict(d):
    d['County'] = titlecase(d['County'])
    d['Station Name'] = titlecase(d['Station Name'])
    d['River Basin'] = titlecase(d['River Basin'])
    
    # numbers
    d['Latitude'] = float(d['Latitude'].strip())
    d['Longitude'] = float(d['Longitude'].strip())
    d['ElevationFeet'] = float(d['ElevationFeet'].strip())
    
    # there is no map
    if 'Map' in d:
        del d['Map']
    return d
```

```python
keyed_collection = [cleanReservoirDict(dict(zip(headings,v))) for v in values]
keyed_collection
```

Now we have to get the data on the sizes of the reservoirs.

```python
reservoir_info_url = "http://cdec.water.ca.gov/misc/resinfo.html"
reservoir_info_doc = Soup(requests.get(reservoir_info_url).content)
```

```python
# get t.findAll('tr')he table on this page
reservoir_info_tr = reservoir_info_doc.find('table').findAll('tr')
# we go for 'font' because their page has an HTML error, I think?
reservoir_info_headers = [[h.text.strip() for h in row.findAll('font')] for row in reservoir_info_tr if row.find('th')][0]
reservoir_info_headers
```

```python
reservoir_info_rows = [[d.text.strip() for d in row.findAll('td')] for row in reservoir_info_tr if row.find('td')]
reservoir_info = [dict(zip(reservoir_info_headers,v)) for v in reservoir_info_rows]
reservoir_info
```
